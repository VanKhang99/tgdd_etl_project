```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager

import time
import re
```

```python
import pandas as pd
```

```python
service = Service(executable_path=ChromeDriverManager().install())
options = Options()

driver = webdriver.Chrome(service=service, options=options)

url = "https://www.thegioididong.com/dong-ho-thong-minh-xiaomi#c=7077&m=17197,17188,17189,19546,17190,20482,18653,18727,36914,19817,18728,17196,36984&o=17&pi=6"

driver.get(url)
time.sleep(7)

parent_tags_list = driver.find_elements(By.XPATH, "//li[@class=' item  __cate_7077']")

def check_element_html(parent_tag, xpath, one_tag=True):
    try:
        return parent_tag.find_element(By.XPATH, xpath) if one_tag else parent_tag.find_elements(By.XPATH, xpath)
    except NoSuchElementException:
        return None
    
def smartwatch_specifications(): 
    products = []
    # i = 0
    for parent_tag in parent_tags_list:
        product_id = parent_tag.get_attribute('data-id')
        face_diameter_mm = None
        screen = None
        health_features_support = None
        exercise_modes_support = None
        notify_apps_support = None
        
        name_tag = check_element_html(parent_tag, "./a[@class='main-contain ']//h3")
        if name_tag is not None:
            title = name_tag.get_attribute('title')
            face_diameter = re.findall(r'[\d.]+(?=\s?mm)', title)
            
            if len(face_diameter) > 0:
                face_diameter_mm = float(face_diameter[0]) 
            
        specification_tags = check_element_html(parent_tag, ".//div[@class='utility']//p", False)
        if specification_tags is not None:
            # len = 2: screen, health_features_support
            # len = 3: screen, health_features_support, notify_apps_support
            # len = 4: screen, health_features_support, exercise_modes_support, notify_apps_support
            
            data = [tag.text for tag in specification_tags]
            
            screen = data[0].split(':')[1].strip()
            health_features_support = data[1].split(':')[1].strip()
    
            if len(data) == 3:
                notify_apps_support = data[2].split(':')[1].strip()
                
            if len(data) == 4:
                exercise_modes_support = data[2].split(':')[1].strip()
                notify_apps_support = data[3].split(':')[1].strip()
   
        products.append([product_id, screen, health_features_support, exercise_modes_support, notify_apps_support])
        # print(f'############ {i} ############')
        # i += 1
       
    df = pd.DataFrame(products, columns=['product_id', 'screen', 'health_features_support', 'exercise_modes_support', 'notify_apps_support'])
    return df 
df = smartwatch_specifications()       
```
