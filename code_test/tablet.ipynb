```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager

import time
import re
```

```python
service = Service(executable_path=ChromeDriverManager().install())
options = Options()

driver = webdriver.Chrome(service=service, options=options)

url = "https://www.thegioididong.com/may-tinh-bang-apple-ipad"

driver.get(url)
time.sleep(7)

parent_tags_list = driver.find_elements(By.XPATH, "//li[@class=' item ajaxed __cate_522']")

def check_element_html(parent_tag, xpath, one_tag=True):
    try:
        return parent_tag.find_element(By.XPATH, xpath) if one_tag else parent_tag.find_elements(By.XPATH, xpath)
    except NoSuchElementException:
        return None
    
def phone_specifications(): 
    for parent_tag in parent_tags_list:
        ### Find tags for multiple tables ###
        
        product_id = parent_tag.get_attribute('data-id')
        product = {'product_id': product_id}
        
        screen_tags = check_element_html(parent_tag, "./a[@class='main-contain ']//div[@class='item-compare gray-bg']/span", False)
        if screen_tags is not None:
            for tag in screen_tags:
                content = tag.text
                
                if '"' in content:
                    product.update({'screen_inch': float(content.replace('"', ''))})
                else:
                    product.update({'screen_type': content.strip()})
        
        specification_tags = check_element_html(parent_tag, ".//div[@class='utility']//p", False)
        if specification_tags is not None:
            for tag in specification_tags:
                content = tag.text
                
                if 'Chip' in content:
                    product.update({'chip': content.strip()})
                elif 'RAM' in content:
                    product.update({'ram': int(re.findall(r'[\d]+', tag.text)[0])})
                elif 'Dung lượng' in content:
                    product.update({'capacity': int(re.findall(r'[\d]+', tag.text)[0])})
                elif 'mAh' or 'Wh' or 'W' in content:
                    data = content.split(', ')
                    # product.update({
                    #     'battery': data[0].strip(), 
                    #     'charge_W': int(re.findall(r'[\d]+', data[1])[0])
                    # }) 
                    print(data) 
                else:
                    product.update({'features_support': content.strip()})
                             
        print(product)
            
phone_specifications()
```
